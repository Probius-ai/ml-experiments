```python
from torchviz import make_dot
import torch
from ultralytics import YOLO

# YOLO 모델 로드
model = YOLO("yolov8n.pt")
model = model.model  # YOLO의 torch.nn.Module 추출

# 더미 입력 생성 (배치 크기 1, 3채널, 640x640 이미지)
dummy_input = torch.randn(1, 3, 640, 640)

# 모델의 forward 실행
output = model(dummy_input)

# 리스트로 반환되면 첫 번째 텐서만 선택
if isinstance(output, (list, tuple)):  
    output = output[0]  # 첫 번째 요소만 선택

# torchviz를 사용하여 모델 구조 시각화
dot = make_dot(output, params=dict(model.named_parameters()))

# 그래프 저장 및 출력
dot.render("yolo_model", format="png")  # PNG 파일로 저장
dot
```
